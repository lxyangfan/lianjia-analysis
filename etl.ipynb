```python
#-*- encoding:utf-8 -*-
import pandas as pd
import json
```

```python
def load_dict_list_json(file_name):
    """
        读取json到dict_list中
    :param file_name:
    :return:
    """
    with open(file_name, mode="r") as json_file:
        return json.load(json_file) 
```

```python
raw_list = load_dict_list_json("data/preowened-2017-06-30.json")
dict_list = []
for one_page_data in raw_list:
    dict_list.extend(one_page_data)
dist_town_price = pd.DataFrame.from_dict(dict_list).astype({'price': float })
dist_town_price
```

```python
preowned = dist_town_price[["addr", "district", "town", "year", "price", "size" ]]
preowned["unit_size"] = preowned["price"] / preowned["size"]
preowned["addr"]  = preowned["district"] +preowned["town"] +preowned["addr"] 
preowned
```

```python
ss = preowned.to_dict(orient="records")
ss
```
